```python
import pandas as pd
import numpy as np
from sklearn.linear_model import Ridge
from sklearn.metrics import make_scorer
from scipy import stats
from sklearn.model_selection import GridSearchCV

import matplotlib.pyplot as plt
```

```python
compound_names = pd.read_csv('processed_data/compound_names.csv',header=None,sep='\n').squeeze()
X_train = pd.read_csv('processed_data/X_train.csv',header=None)
Y_train = pd.read_csv('processed_data/Y_train.csv',header=None)
X_test = pd.read_csv('processed_data/X_test.csv',header=None)
Y_test = pd.read_csv('processed_data/Y_test.csv',header=None)
Y_pred_mNODE = pd.read_csv('results/predicted_metabolomic_profiles.csv',header=None)
```

```python
def mean_rho_test(y_t,y_p):
    rhovec = []
    for k in range(len(y_p.T)):
        rho,p = stats.spearmanr(y_p[:,k],y_t[:,k])
        rhovec.append(rho)
    return np.asarray(rhovec).mean()
mean_rho_score = make_scorer(mean_rho_test)

alpha_grid = {'alpha':np.linspace(0.1,1,10)}
ridge = Ridge(alpha=.5,max_iter=100000,fit_intercept=True)
ridge_opt = GridSearchCV(ridge,alpha_grid,scoring=mean_rho_score,cv=5,n_jobs=-1)
out=ridge_opt.fit(X_train.values,Y_train.values)
```

```python
Y_pred = pd.DataFrame(ridge_opt.predict(X_test))
rhovec = Y_pred.corrwith(Y_test,method='spearman').values
#rhos_valid = rhovec[~compound_names.isna().values.squeeze()]
#rhos_valid.sort()
rhovec.sort()
score_1 = rhovec.mean()
score_2 = rhovec[-50:].mean()
score_3 = (rhovec>0.5).sum()

print('The mean Spearman C.C. for all metabolites: '+str(score_1))
print('Top 50 Spearman C.C.: '+str(score_2))
print('The number of metabolites with Spearman C.C. > 0.5: '+str(score_3))
```

```python
plt.scatter(Y_test.values[:,~compound_names.isna().values.squeeze()],Y_pred.values[:,~compound_names.isna().values.squeeze()],alpha=0.01)
plt.gca().set_xlabel('True metabolite level')
plt.gca().set_ylabel('Predicted metabolite level (Ridge)')
plt.show()
```

```python
plt.scatter(Y_test.values[:,~compound_names.isna().values.squeeze()],Y_pred_mNODE.values[:,~compound_names.isna().values.squeeze()],alpha=0.01)
plt.gca().set_xlabel('True metabolite level')
plt.gca().set_ylabel('Predicted metabolite level (mNODE)')
plt.show()
```

```python
plt.scatter(Y_pred_mNODE.values[:,~compound_names.isna().values.squeeze()],Y_pred.values[:,~compound_names.isna().values.squeeze()],alpha=0.01)
plt.plot([-4,7],[-4,7],'k')
plt.gca().set_xlabel('mNODE prediction')
plt.gca().set_ylabel('Ridge prediction')
plt.show()
```

```python
rhovec_ridge = Y_pred.corrwith(Y_test,method='spearman').values
rhovec_mNODE = Y_pred_mNODE.corrwith(Y_test,method='spearman').values
plt.scatter(rhovec_mNODE[~compound_names.isna().values.squeeze()],rhovec_ridge[~compound_names.isna().values.squeeze()])
plt.plot([-.2,.8],[-.2,.8],'k')
plt.gca().set_xlabel(r'$\rho$ mNODE')
plt.gca().set_ylabel(r'$\rho$ Ridge')
plt.show()
```
